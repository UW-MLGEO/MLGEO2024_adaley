## Declaring Python Function

```python
def is_prime(n):
    """Check if a number is a prime number."""
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True

# Example usage:
print(is_prime(11))  # True
print(is_prime(4))   # False
```

## Plotting

```python
gridsize = (1, 2)
fig = plt.figure(figsize=(10,4))

#Scatter plot
ax1 = plt.subplot2grid(gridsize, (0,0), colspan=1, rowspan=1)
ax1.scatter(float_array[:,0], float_array[:,1], label='scatter points') #Plot the array
ax1.legend(loc='center left')
ax1.grid(alpha=0.5)
ax1.set_xlabel('x-axis (units)')
ax1.set_ylabel('y-axis (units)')
ax1.set_title('Scatter plot of random floats')

#Bar plot
ax2 = plt.subplot2grid(gridsize, (0,1), colspan=1, rowspan=1)
ax2.bar(month_list, bar_data, label='bar data') #Plot the list
ax2.legend()
ax2.grid(alpha=0.5)
ax2.set_title('Random Bar Plot')
ax2.set_xlabel('Months')
ax2.set_ylabel('Value')

plt.savefig(PNG + 'bar_and_scatter_plot.png', dpi=300, bbox_inches='tight',
                facecolor='w', transparent=False)
```
